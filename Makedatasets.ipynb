```python
#coarse-scaleup, down
for num in range(30,50):
    bckimg = cv.resize(cv.imread('datasets/mirflickr/im'+str(870+num)+'.jpg', 
                                 cv.IMREAD_GRAYSCALE), dsize=(256, 256), interpolation=cv.INTER_AREA)
    num0 = cv.resize(cv.imread('datasets/MNIST/mnist_train/train/'+str(file_list[num+2060]), cv.IMREAD_UNCHANGED), 
                     dsize=(200, 200), interpolation = cv.INTER_AREA)
    mxpivot=float("inf")
    pxpivot=float("-inf")
    mypivot=float("inf")
    pypivot=float("-inf")
    
    for y in range(200):
        for z in range(200):
            if num0[y][z] > 30:
                if z<mxpivot: mxpivot=z
                if z>pxpivot: pxpivot=z
                if y<mypivot: mypivot=y
                if y>pypivot: pypivot=y
                    
    a = [64, 200]+[random.randrange(64,201) for value in range(0, 78)]
    a.sort(reverse=True)
    xp=random.randrange(-mypivot, 250-pypivot)
    yp=random.randrange(-mxpivot, 250-pxpivot)
    
    for x in range(80):
        img = bckimg
        img=img+1
        img=img-1
        numk = cv.resize(num0, dsize=(a[x],a[x]), interpolation=cv.INTER_AREA)
        newxp=100+xp-int(a[x]/2)
        newyp=100+yp-int(a[x]/2)
        for y in range(a[x]):
            for z in range(a[x]):            
                if numk[y][z] > 30:
                    img[y+newxp][z+newyp]=numk[y][z]
        cv.imwrite('datasets/trimmed/coarse/9/scaledown/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
```

```python
#coarse-rotate
for num in range(30, 50):
    a=random.randrange(64,201)
    bckimg = cv.resize(cv.imread('datasets/mirflickr/im'+str(20570+num)+'.jpg', 
                                 cv.IMREAD_GRAYSCALE), dsize=(256, 256), interpolation=cv.INTER_AREA)
    num0 = cv.resize(cv.imread('datasets/MNIST/mnist_train/train/'+str(file_list[num+720]), 
                               cv.IMREAD_UNCHANGED), dsize=(a, a), interpolation = cv.INTER_AREA)
    degree=[0,360]+[random.randrange(0,360) for value in range(78)]
    degree.sort(reverse=True)
    mxpivot=float("inf")
    pxpivot=float("-inf")
    mypivot=float("inf")
    pypivot=float("-inf")
    for y in range(a):
        for z in range(a):
            if num0[y][z] > 30:
                if z<mxpivot: mxpivot=z
                if z>pxpivot: pxpivot=z
                if y<mypivot: mypivot=y
                if y>pypivot: pypivot=y
                    
    pivot=min(mxpivot, a-pxpivot,mypivot,a-pypivot)
    xp=random.randrange(-pivot, 256-a)
    yp=random.randrange(-pivot, 256-a)

    for x in range(80):
        img = bckimg
        img=img+1
        img=img-1
        rotation=cv.getRotationMatrix2D((int(a/2),int(a/2)),degree[x],1)
        numk=cv.warpAffine(num0, rotation, (a,a))
        for y in range(a):
            for z in range(a):            
                if numk[y][z] > 30:
                    img[y+xp][z+yp]=numk[y][z]
        cv.imwrite('datasets/trimmed/coarse/9/rotatecw/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
```

```python
#fine-scaleup, down
for k in range(10):
    file_list=[file for file in os.listdir('datasets/MNIST/mnist_train/train') if file.endswith('num'+str(k)+'.png')]
    for num in range(50):
        f_start=random.randrange(0,60)
        f_end=random.randrange(f_start+10,min(f_start+20,80))

        json_object={}
        with open("datasets/trimmed/fine/label.json", 'r') as f:
            json_object = json.load(f)

        json_object[str(k)]["scaleup"]["f_start"].append(f_start)
        json_object[str(k)]["scaleup"]["f_end"].append(f_end)
        with open('datasets/trimmed/fine/label.json', 'w') as f:
            json.dump(json_object, f) 

        a_start=random.randrange(64,160)
        a_end=random.randrange(a_start+10,201)
        a = [a_start, a_end]+[random.randrange(a_start,a_end) for value in range(0, f_end-f_start+1)]
        a.sort()

        bckimg = cv.resize(cv.imread('datasets/mirflickr/im'+str(3000+k*50+num)+'.jpg', 
                                     cv.IMREAD_GRAYSCALE), dsize=(256, 256), interpolation=cv.INTER_AREA)
        num0 = cv.resize(cv.imread('datasets/MNIST/mnist_train/train/'+str(file_list[num+2300]), cv.IMREAD_UNCHANGED), 
                         dsize=(a_end, a_end), interpolation = cv.INTER_AREA)
        mxpivot=float("inf")
        pxpivot=float("-inf")
        mypivot=float("inf")
        pypivot=float("-inf")

        for y in range(a_end):
            for z in range(a_end):
                if num0[y][z] > 30:
                    if z<mxpivot: mxpivot=z
                    if z>pxpivot: pxpivot=z
                    if y<mypivot: mypivot=y
                    if y>pypivot: pypivot=y

        xp=random.randrange(-mypivot, 250-pypivot)
        yp=random.randrange(-mxpivot, 250-pxpivot)

        for x in range(80):
            img = bckimg
            img=img+1
            img=img-1
            if x<f_start:
                numk = cv.resize(num0, dsize=(a_start,a_start), interpolation=cv.INTER_AREA)
                newxp=int(a_end/2)+xp-int(a_start/2)
                newyp=int(a_end/2)+yp-int(a_start/2)
                for y in range(a_start):
                    for z in range(a_start):            
                        if numk[y][z] > 30:
                            img[y+newxp][z+newyp]=numk[y][z]
                cv.imwrite('datasets/trimmed/fine/'+str(k)+'/scaleup/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
            elif x<f_end:
                numk = cv.resize(num0, dsize=(a[x-f_start],a[x-f_start]), interpolation=cv.INTER_AREA)
                newxp=int(a_end/2)+xp-int(a[x-f_start]/2)
                newyp=int(a_end/2)+yp-int(a[x-f_start]/2)
                for y in range(a[x-f_start]):
                    for z in range(a[x-f_start]):            
                        if numk[y][z] > 30:
                            img[y+newxp][z+newyp]=numk[y][z]
                cv.imwrite('datasets/trimmed/fine/'+str(k)+'/scaleup/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
            else:
                for y in range(a_end):
                    for z in range(a_end):            
                        if num0[y][z] > 30:
                            img[y+xp][z+yp]=num0[y][z]
                cv.imwrite('datasets/trimmed/fine/'+str(k)+'/scaleup/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
```

```python
#fine-rotate
for k in range(9,10):
    file_list=[file for file in os.listdir('datasets/MNIST/mnist_train/train') if file.endswith('num'+str(k)+'.png')]
    for num in range(50):
        f_start=random.randrange(0,60)
        f_end=random.randrange(f_start+10,min(f_start+20,80))
        a=random.randrange(64,201)
    
        json_object={}
        with open("datasets/trimmed/fine/label.json", 'r') as f:
            json_object = json.load(f)
        
        json_object[str(k)]["rotateccw"]["f_start"].append(f_start)
        json_object[str(k)]["rotateccw"]["f_end"].append(f_end)
        with open('datasets/trimmed/fine/label.json', 'w') as f:
            json.dump(json_object, f)
    
        bckimg = cv.resize(cv.imread('datasets/mirflickr/im'+str(3500+50*k+num)+'.jpg', 
                                 cv.IMREAD_GRAYSCALE), dsize=(256, 256), interpolation=cv.INTER_AREA)
        num0 = cv.resize(cv.imread('datasets/MNIST/mnist_train/train/'+str(file_list[num+2350]), 
                               cv.IMREAD_UNCHANGED), dsize=(a, a), interpolation = cv.INTER_AREA)
    
        deg_start=random.randrange(0,300)
        deg_end=random.randrange(deg_start+20,361)
        degree=[deg_start, deg_end]+[random.randrange(deg_start, deg_end) for value in range(0, f_end-f_start+1)]
        degree.sort()
    
        mxpivot=float("inf")
        pxpivot=float("-inf")
        mypivot=float("inf")
        pypivot=float("-inf")
        for y in range(a):
            for z in range(a):
                if num0[y][z] > 30:
                    if z<mxpivot: mxpivot=z
                    if z>pxpivot: pxpivot=z
                    if y<mypivot: mypivot=y
                    if y>pypivot: pypivot=y
                    
        pivot=min(mxpivot, a-pxpivot,mypivot,a-pypivot)
        xp=random.randrange(-pivot, 256-a)
        yp=random.randrange(-pivot, 256-a)

        for x in range(80):
            img = bckimg
            img=img+1
            img=img-1
            if x<f_start:
                rotation=cv.getRotationMatrix2D((int(a/2),int(a/2)),deg_start,1)
                numk=cv.warpAffine(num0, rotation, (a,a))
                for y in range(a):
                    for z in range(a):
                        if numk[y][z]>30:
                            img[y+xp][z+yp]=numk[y][z]
                cv.imwrite('datasets/trimmed/fine/'+str(k)+'/rotateccw/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
            elif x<f_end:
                rotation=cv.getRotationMatrix2D((int(a/2),int(a/2)),degree[x-f_start],1)
                numk=cv.warpAffine(num0, rotation, (a,a))
                for y in range(a):
                    for z in range(a):            
                        if numk[y][z] > 30:
                            img[y+xp][z+yp]=numk[y][z]
                cv.imwrite('datasets/trimmed/fine/'+str(k)+'/rotateccw/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
            else:
                rotation=cv.getRotationMatrix2D((int(a/2),int(a/2)),deg_end,1)
                numk=cv.warpAffine(num0, rotation, (a,a))
                for y in range(a):
                    for z in range(a):
                        if numk[y][z]>30:
                            img[y+xp][z+yp]=numk[y][z]
                cv.imwrite('datasets/trimmed/fine/'+str(k)+'/rotateccw/'+str(num)+'/image_'+str(x).zfill(2)+'.jpg',img)
```
